```python
def scoring (a,b):
    a_tag_sayisi = a[1]
    b_tag_sayisi = b[1]
    count = 0
    for x in range(2,a_tag_sayisi):
        for y in range(2,b_tag_sayisi):
            if(a[x] == b[y]):
                count += 1
    return min(a_tag_sayisi - count,b_tag_sayisi - count, count)            
```
